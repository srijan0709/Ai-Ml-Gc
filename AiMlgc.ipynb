```python
df_0 = newdf[newdf['MEL'] == 1]
df_1 = newdf[newdf['NV'] == 1]
df_2 = newdf[newdf['BCC'] == 1]
df_3 = newdf[newdf['AKIEC'] == 1]
df_4 = newdf[newdf['BKL'] == 1]
df_5 = newdf[newdf['DF'] == 1]
df_6 = newdf[newdf['VASC'] == 1]

from sklearn.utils import resample
n_samples=500
df_0_balanced = resample(df_0, replace=True, n_samples=n_samples, random_state=42)
df_1_balanced = resample(df_1, replace=True, n_samples=n_samples, random_state=42)
df_2_balanced = resample(df_2, replace=True, n_samples=n_samples, random_state=42)
df_3_balanced = resample(df_3, replace=True, n_samples=n_samples, random_state=42)
df_4_balanced = resample(df_4, replace=True, n_samples=n_samples, random_state=42)
df_5_balanced = resample(df_5, replace=True, n_samples=n_samples, random_state=42)
df_6_balanced = resample(df_6, replace=True, n_samples=n_samples, random_state=42)
df_balanced = pd.concat([df_0_balanced, df_1_balanced,
                              df_2_balanced, df_3_balanced,
                              df_4_balanced, df_5_balanced, df_6_balanced])
df_balanced
```

```python
selected_columns = ['MEL', 'NV', 'BCC','AKIEC','BKL','DF','VASC']  # List of columns you want to select
Y_df = df_balanced.loc[:, selected_columns]
Y_df
```

```python
## Now we create dataloaders and dataet for the training resnet
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import numpy as np
class CustomDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        img = Image.open(img_path).convert("RGB")

        if self.transform:
            img = self.transform(img)

        label = self.labels[idx]

        return img, label
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize to desired size
    transforms.ToTensor(),           # Convert PIL Image to tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Normalize
])
```

```python
image_paths = df_balanced['path']
image_paths = image_paths.tolist()
# Assuming labels are already one-hot encoded, for example, 3 classes
labels = np.array(Y_df)

labels = torch.tensor(labels)
```

```python
dataset = CustomDataset(image_paths, labels, transform=transform)
batch_size = 32
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
## Now we will building the testing dataset
path = "/content/KCDH2024_Test_Input/KCDH2024_Test_Input/"
dir_list = os.listdir(path)
list_test_id = []
list_test_path = []
for elem in dir_list:
  if(elem=="LICENSE.txt"):
      continue
  else:
      list_test_path.append(path+elem)
      list_test_id.append(elem[0:-4])
```

```python
class TestDataset(Dataset):
    def __init__(self, image_paths, transform=None):
        self.image_paths = image_paths
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        image = Image.open(image_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        return image

# Set your test data directory

test_image_paths = list_test_path  # List of test image directory paths

# Define transformations for test data
test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Create test dataset and dataloader
test_dataset = TestDataset(test_image_paths, transform=test_transform)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
## now we will build the val dataset

val_image_path = image_paths[0:10]
# val_image_path = val_image_path.tolist()
# val_labels = np.array(Y_df[0:10])

# val_labels = torch.tensor(val_labels)

# val_dataset = CustomDataset(val_image_path, val_labels, transform=transform)
# batch_size = 32
# val_dataloader = DataLoader(val_dataset, batch_size=batch_size, shuffle=True)

val_dataset = TestDataset(val_image_path, transform=test_transform)
val_dataloader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

```python
## now we will build resnet for image classification
import torch
import torch.nn as nn
import torchvision.models as models
import torch.optim as optim
```

```python
class ResNetModel(nn.Module):
    def __init__(self, num_classes):
        super(ResNetModel, self).__init__()
        self.resnet = models.resnet50(pretrained=True)
        num_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Linear(num_features, num_classes)

    def forward(self, x):
        return self.resnet(x)
```

```python
num_classes = 7  # Change this according to your number of classes
model = ResNetModel(num_classes)
```

```python
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 42
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
print(device)
```

```python
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for index , (images, labels) in enumerate(dataloader):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels.float())
        loss.backward()
        optimizer.step()
        _, temp_labels = torch.max(labels, 1)
        running_loss += loss.item() * images.size(0)
        print(running_loss /((index + 1) * 32) )
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == temp_labels).sum().item()


    epoch_accuracy = 100 * correct / total
    epoch_loss = running_loss / len(dataset)
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%')


```
