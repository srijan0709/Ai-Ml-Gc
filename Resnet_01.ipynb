```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import numpy as np
class TrainData(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        img = Image.open(img_path).convert("RGB")

        if self.transform:
            img = self.transform(img)

        label = self.labels[idx]

        return img, label



class ValData(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        img = Image.open(img_path).convert("RGB")

        if self.transform:
            img = self.transform(img)

        label = self.labels[idx]

        return img, label
```

```python
# class ResNetModel(nn.Module):
#     def __init__(self, num_classes):
#         super(ResNetModel, self).__init__()
#         self.resnet = models.resnet50(pretrained=True)
#         num_features = self.resnet.fc.in_features
#         self.resnet.fc = nn.Linear(num_features, num_classes)

#     def forward(self, x):
#         return self.resnet(x)
import torch.nn as nn
import torch.nn.functional as F

# Freeze initial layers
# for param in model.parameters():
#     param.requires_grad = False
# num_classes =7
# Modify the last fully connected layer for your classification task
# num_ftrs = model.fc.in_features
# model.fc1 = nn.Linear(512, 256)
# model.fc2 = nn.Linear(256, 128)
# model.fc3 = nn.Linear(128, 64)
# model.fc4 = nn.Linear(64, 512)
# model.fc5 = nn.Linear(512, 256)
# model.fc6 = nn.Linear(256, 128)
# model.fc7 = nn.Linear(128, 64)
# model.fc8 = nn.Linear(64, 32)
# model.fc9 = nn.Linear(32, 16)
# model.fc10 = nn.Linear(16, 7)


import torch.nn.functional as F
import torch.nn as nn
import torchvision.models as models

class ResNetModel(nn.Module):
    def __init__(self, num_classes):
        super(ResNetModel, self).__init__()
        self.resnet = models.resnet18(pretrained=True)
        for param in self.resnet.parameters():
            param.requires_grad = False
        # num_ftrs = self.resnet.fc.in_features
        self.fc0 = nn.Linear(1000, 512)
        self.fc1 = nn.Linear(512, 256)
        self.dropout1 = nn.Dropout(p=0.2)  # Changed dropout rate to 0.2
        self.fc2 = nn.Linear(256, 128)
        self.dropout2 = nn.Dropout(p=0.2)  # Changed dropout rate to 0.2
        self.fc3 = nn.Linear(128, 64)
        self.dropout3 = nn.Dropout(p=0.2)  # Changed dropout rate to 0.2
        self.fc4 = nn.Linear(64, 512)
        self.dropout4 = nn.Dropout(p=0.2)  # Changed dropout rate to 0.2
        self.fc5 = nn.Linear(512, 256)
        self.dropout5 = nn.Dropout(p=0.2)  # Changed dropout rate to 0.2
        self.fc6 = nn.Linear(256, 128)
        self.dropout6 = nn.Dropout(p=0.2)  # Changed dropout rate to 0.2
        self.fc7 = nn.Linear(128, 64)
        self.dropout7 = nn.Dropout(p=0.2)  # Changed dropout rate to 0.2
        self.fc8 = nn.Linear(64, 32)
        self.dropout8 = nn.Dropout(p=0.2)  # Changed dropout rate to 0.2
        self.fc9 = nn.Linear(32, 16)
        self.dropout9 = nn.Dropout(p=0.2)  # Changed dropout rate to 0.2
        self.fc10 = nn.Linear(16, 7)

    def forward(self, x):
        x = self.fc0(self.resnet(x))
        x = F.relu(self.fc1(x))
        x = self.dropout1(x)
        x = F.relu(self.fc2(x))
        x = self.dropout2(x)
        x = F.relu(self.fc3(x))
        x = self.dropout3(x)
        x = F.relu(self.fc4(x))
        x = self.dropout4(x)
        x = F.relu(self.fc5(x))
        x = self.dropout5(x)
        x = F.relu(self.fc6(x))
        x = self.dropout6(x)
        x = F.relu(self.fc7(x))
        x = self.dropout7(x)
        x = F.relu(self.fc8(x))
        x = self.dropout8(x)
        x = F.relu(self.fc9(x))
        x = self.dropout9(x)
        x = self.fc10(x)
        return x
```

```python
num_classes = 7

# Create the model
model = ResNetModel(num_classes=num_classes)
```

```python
from torchsummary import summary

# Use GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
print(device)

 # Assuming input channels first

# Print the model summary
summary(model, (3,299,299))
```

```python
from tqdm import tqdm
# model = ResNetModel(num_classes=num_classes).to('cuda')
model = ResNetModel(num_classes=num_classes)
model.load_state_dict(torch.load(r"C:\Users\srija\Assignment\AI_ML_GC\resnet_model_49.pth", map_location=torch.device('cuda')))  # Load the model
model.to('cuda')

# model.load_state_dict(torch.load('/content/resnet_model_8.pth', map_location=torch.device('cuda')))
# Set model to training mode
model.train()
learning_rate = 0.0002

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
num_epochs = 50

# Training loop
for epoch in range(num_epochs):
    running_loss = 0.0
    correct = 0
    total = 0

    # Wrap the dataloader with tqdm for progress bar
    for images, labels in tqdm(train_dataloader, desc=f'Epoch {epoch+1}/{num_epochs}'):
        optimizer.zero_grad()
        images = images.to('cuda')
        labels = labels.to('cuda')
        outputs = model(images)
        _, temp_labels = torch.max(labels, 1)
        loss = criterion(outputs, labels.float())
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == temp_labels).sum().item()

    train_loss = running_loss / len(train_dataloader)
    train_accuracy = 100 * correct / total

    model.eval()  # Set the model to evaluation mode
    correct_val = 0
    total_val = 0
    val_running_loss = 0.0

    with torch.no_grad():
        for images, labels in tqdm(val_dataloader, desc=f'Validation'):
            images = images.to('cuda')
            labels = labels.to('cuda')
            outputs = model(images)
            _, temp_labels = torch.max(labels, 1)
            val_loss = criterion(outputs, labels.float())
            val_running_loss += val_loss.item()

            _, predicted = torch.max(outputs, 1)
            total_val += labels.size(0)
            correct_val += (predicted == temp_labels).sum().item()

    # Calculate validation accuracy
    val_loss = val_running_loss / len(val_dataloader)
    val_accuracy = 100 * correct_val / total_val

    # Save model checkpoint
    torch.save(model.state_dict(), f'resnet_model_{epoch+50}.pth')
    # files.download(f'resnet_model_{epoch}.pth')
    print(f'Epoch [{epoch+1}/{num_epochs}], Training Loss: {train_loss:.4f}, Training Accuracy: {train_accuracy:.2f}%, Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.2f}%')
```
